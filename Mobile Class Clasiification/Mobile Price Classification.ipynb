Let's get to the Model Selection phase

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X=train.iloc[:,:-1]
Y=train.iloc[:,-1]
sc=StandardScaler()

X=sc.fit_transform(X)

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

log=LogisticRegression()

log.fit(X_train,Y_train)

print('Training Accuracy of Logistic Regression is : {}'.format(log.score(X_train,Y_train)*100))

y_predlog=log.predict(X_test)

acc_log=accuracy_score(Y_test,y_predlog)*100

print("Accuracy of Logistic Regression Classifier is: {}%".format(acc_log))

print("Confusion Matrix of Logistic Regression Classifier is: \n{}".format(confusion_matrix(Y_test, y_predlog)))

print("{}".format(classification_report(Y_test, y_predlog)))
```

SVM

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'rbf')

svc.fit(X_train, Y_train)

print("Traiing Score of SVM is: {}".format(svc.score(X_train, Y_train) * 100))

y_predsvm = svc.predict(X_test)

acc_svm = accuracy_score(Y_test, y_predsvm) * 100

print("Accuracy of SVM Classifier is: {}%".format(acc_svm))

print("Confusion Matrix of SVM Classifier is: \n{}".format(confusion_matrix(Y_test, y_predsvm)))

print("{}".format(classification_report(Y_test, y_predsvm)))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion = 'entropy')

dt.fit(X_train, Y_train)
print("Traiing Score of Decision Tree is: {}".format(dt.score(X_train, Y_train) * 100))

y_predtree = dt.predict(X_test)

acc_dt = accuracy_score(Y_test, y_predtree) * 100
print("Accuracy of SVM Classifier is: {}%".format(acc_dt))
print("Confusion Matrix of SVM Classifier is: \n{}".format(confusion_matrix(Y_test, y_predtree)))
print("{}".format(classification_report(Y_test, y_predtree)))
```

Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
train_accuracy = []
test_accuracy = []

for i in range(100, 600, 100):
    c = RandomForestRegressor(n_estimators = i)
    c.fit(X_train, Y_train)
    train_accuracy.append(c.score(X_train, Y_train))
    test_accuracy.append(c.score(X_test, Y_test))
    
frame_rf = pd.DataFrame({"n_neighbors": range(100, 600, 100), "train_accuracy": train_accuracy, "test_accuracy": test_accuracy})
frame_rf
```

Compare between accuracy of every model

```python
plt.figure(figsize=(15,10))
classifiers = ['Logistic Regression',  'SVM', 'Decision Tree','Random Foresgt Regressor']
lst_acc = [acc_log, acc_svm, acc_dt,frame_rf['test_accuracy'].max()*100]
df = pd.DataFrame({'Model': classifiers, 'Accuracy': lst_acc})
sns.barplot(data = df, x = 'Model', y = 'Accuracy')
```

Let's Try using folds

```python
from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score

kfolds=KFold(n_splits=10)

cv_mean=[]

accuracy=[]

classifires=['Logistic Regression','SVM','Decision Tree']

models=[LogisticRegression(),SVC(kernel='rbf'),DecisionTreeClassifier()]

for i in models:
    
    model=i
    
    cv_result=cross_val_score(model,X_train,Y_train,cv=kfolds,scoring='accuracy')
    
    cv_mean.append(cv_result.mean())
    
    accuracy.append(cv_result)


cv_models=pd.DataFrame(cv_mean, index=classifires)

cv_models.columns=['CV Mean']

cv_models.sort_values(['CV Mean'], ascending=[0])    
```

Logistic Regression Got the highest accuracy
